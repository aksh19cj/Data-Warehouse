### Compare the sales between wine & and beer during Thanksgiving?

#Thanksgiving - week 11 (1989), week 63(1990), Week 116(1991), Week 168 ( 1992), week 220(1993), Week 272(1994), Week 324(1995), Week 377(1996)

```python
Weeks_ccount_data = df[df['WEEK'].isin([11,63,116,168,220,272,324,377])]
```

```python
beer_wine = Weeks_ccount_data[['WEEK', 'BEER', 'WINE']]
```

```python
Thanksgiving_yearly_sales = beer_wine.groupby('WEEK').sum()
```

```python
year = [1989,1990,1991,1992,1993,1994,1995,1996]
```

```python
Thanksgiving_yearly_sales['YEAR'] = year
```

```python
Thanksgiving_yearly_sales
```

```python
Thanksgiving_yearly_sales_1 = Thanksgiving_yearly_sales[['BEER','YEAR']]
Thanksgiving_yearly_sales_2 = Thanksgiving_yearly_sales[['WINE','YEAR']]
Thanksgiving_yearly_sales_1['Category'] = 'Beer'
Thanksgiving_yearly_sales_2['Category'] = 'Wine'
Thanksgiving_yearly_sales_1.rename(columns={"BEER": "Sales"},inplace=True)
Thanksgiving_yearly_sales_2.rename(columns={"WINE": "Sales"},inplace=True)
result = pd.concat([Thanksgiving_yearly_sales_1,Thanksgiving_yearly_sales_2])
plt.figure(figsize=(15,10))
sns.barplot(data=result, x="YEAR", y='Sales',hue='Category')
```
