```python
demograph = demo.dropna()
demo_chicago = demograph[demograph['CITY'] == 'CHICAGO']
demo_reduced = demo_chicago[['CITY',"STORE", 'ZONE']]
demo_reduced['CITY'] = demo_reduced['CITY'].astype('str')
demo_reduced['STORE'] = demo_reduced['STORE'].astype('int')
demo_reduced['ZONE'] = demo_reduced['ZONE'].astype('int')
demo_reduced
```

```python
result = pd.merge(df_chicago,demo_reduced,on ='STORE')
result
```

```python
result_1 = result[['FISH', 'MEATFROZ', 'MEAT', 'ZONE', 'CITY']]
result_1['FISH'] =result_1['FISH'].astype('float') 
result_1['MEAT'] =result_1['MEAT'].astype('float') 
result_1['MEATFROZ'] =result_1['MEATFROZ'].astype('float') 
```

```python
grouped_multiple = result_1.groupby(['ZONE']).agg({'FISH': ['sum'], 'MEAT':'sum', 'MEATFROZ':'sum'})
grouped_multiple.reset_index(inplace= True)
grouped_multiple
```

```python
grouped_multiple_1 = grouped_multiple[['FISH','ZONE']]
grouped_multiple_1
```

```python
grouped_multiple_3 = grouped_multiple[['MEATFROZ','ZONE']]
grouped_multiple_3
```

```python
grouped_multiple_1 = grouped_multiple[['FISH','ZONE']]
grouped_multiple_2 = grouped_multiple[['MEAT','ZONE']]
grouped_multiple_3 = grouped_multiple[['MEATFROZ','ZONE']]
grouped_multiple_1['PRODUCT'] = 'FISH'
grouped_multiple_2['PRODUCT'] = 'MEAT'
grouped_multiple_3['PRODUCT'] = 'MEATFROZ'

grouped_multiple_1.rename(columns={"FISH": "Sales"},inplace=True)
grouped_multiple_2.rename(columns={"MEAT": "Sales"},inplace=True)
grouped_multiple_3.rename(columns={"MEATFROZ": "Sales"},inplace=True)
resulted = pd.concat([grouped_multiple_1,grouped_multiple_2, grouped_multiple_3])
resulted.columns =['SALES', 'ZONE', 'PRODUCT']
resulted
```

```python
plt.figure(figsize=(30,15))
sns.factorplot(data=resulted, x="ZONE", y='SALES',hue='PRODUCT')
```
